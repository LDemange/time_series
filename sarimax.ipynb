# Test et visualisation d'un modèle SARIMAX

```python
date_start_test = '2013-08-02'
N_test=15
N_train=150
scenario='with_exog'
alpha=0.05
```

```python
date_end_test, date_start_train, date_end_train, X_train_endo, X_train_exo, X_test_exo, y_true = select_data(date_start_test,N_test,N_train,scenario)
```

```python
models = pm.auto_arima(y=X_train_endo,
                       X=X_train_exo,
                       seasonal = True, m=7, 
                       start_p=0, max_p=3, start_d = 0, max_d=2, start_q=0, max_q=3,
                       start_P=0, max_P=3, max_D=2, start_Q=0, max_Q=3,
                       stepwise=True, 
                       trace=True, suppress_warnings=True,return_valid_fits=True,test='adf')


```

```python
#Pour tester une liste de models
MAPE_best=999
y_pred_best=0
upper_pred_best=0
lower_pred_best=0
for model in models:
    model_sarimax = SARIMAX(endog=X_train_endo, exog=X_train_exo,
                        order=model.order, seasonal_order=model.seasonal_order) 
    model_sarimax = model_sarimax.fit(start_params=None,maxiter=300,disp=0)
    data_pred = model_sarimax.get_prediction(start=date_start_test, end=date_end_test,exog=X_test_exo).summary_frame(alpha=alpha) #alpha=0.05 docn intervalle de confiance à 95%
    y_pred=data_pred['mean']
    upper_pred=data_pred['mean_ci_upper']
    lower_pred=data_pred['mean_ci_lower']
    MAPE = mean_absolute_percentage_error(y_true, y_pred)
    if(MAPE<=MAPE_best):
        MAPE_best=MAPE
        model_best=model
        y_pred_best=y_pred
        lower_pred_best=lower_pred
        upper_pred_best=upper_pred
        
```

```python
plt.figure(figsize=(15,11))
y_pred_best.plot(color = 'green')
X_train_endo.plot(color = 'blue')
y_true.plot(color = 'red')
plt.fill_between(y_pred_best.index, 
                 lower_pred_best, 
                 upper_pred_best, 
                 color='b', alpha=0.15)

MAPE_MODEL = mean_absolute_percentage_error(y_true, y_pred_best) # MAE en pourcentage 
y_baseline = np.ones(N_test)*X_train_endo[-1]
MAPE_baseline = mean_absolute_percentage_error(y_true, y_baseline) # MAE en pourcentage
print(f'MAPE_MODEL = {MAPE_MODEL}, MAPE_baseline = {MAPE_baseline}')
```

```python
for i in range(0,10):
    date_start_test= pd.to_datetime(date_start_test)+timedelta(days=165)
    date_end_test, date_start_train, date_end_train, X_train_endo, X_train_exo, X_test_exo, y_true = select_data(date_start_test,N_test,N_train,scenario)
    model_sarimax = SARIMAX(endog=X_train_endo, exog=X_train_exo,order=model_best.order, seasonal_order=model_best.seasonal_order) 
    model_sarimax = model_sarimax.fit(start_params=None,maxiter=300,disp=0)
    data_pred = model_sarimax.get_prediction(start=date_start_test, end=date_end_test,exog=X_test_exo).summary_frame(alpha=alpha) #alpha=0.05 docn intervalle de confiance à 95%
    y_pred_best=data_pred['mean']
    MAPE_MODEL = mean_absolute_percentage_error(y_true, y_pred_best) # MAE en pourcentage 
    print(f'MAPE_MODEL = {MAPE_MODEL}, MAPE_baseline = {MAPE_baseline}')
```
